```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/Online_Retail.csv", encoding='ISO-8859-1')
print(df.head())
print(df.info())
```

```python
# حذف مشتریان بدون CustomerID
df = df[df['CustomerID'].notnull()]

# حذف تراکنش‌های با Quantity یا UnitPrice صفر یا منفی
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# بررسی تعداد رکوردها بعد از پاکسازی
print(df.shape)
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# ذخیره برای استفاده در Notebook های بعدی
df.to_csv("../data/clean/retail_clean.csv", index=False)

# بررسی نهایی
print(df.head())
```
